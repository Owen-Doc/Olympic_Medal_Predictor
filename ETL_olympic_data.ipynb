```python
# Dependencies
import pandas as pd

# Files to load
data_host = "data/host_cities.csv"
data_gdp = "data/noc_gdp.csv"
```

```python
# Create Host Cities DataFrame
host_df = pd.read_csv(data_host)
print(host_df.shape)
host_df.head()
```

```python
# Create Olympic Athletes DataFrame
athletes_df = pd.read_csv(data_athletes)
print(athletes_df.shape)
athletes_df.head(20)
```

```python
# Create NOC Regions DataFrame
regions_df = pd.read_csv(data_regions)
print(regions_df.shape)
regions_df.head()
```

```python
# Drop "notes" column
regions_df = regions_df.drop(columns=["notes"])
regions_df.head()
```

```python
# Merge athletes and host citiy dataframes and create new dataframe
olympics_df = athletes_df.merge(host_df, left_on="City", right_on="Host City")
olympics_df = olympics_df.merge(regions_df, on="NOC")
olympics_df.head()
```

```python
# Check for null values
olympics_df.isnull().sum()
```

```python
null = olympics_df.isnull()
null_rows = null.any(axis='columns')
olympics_df[null_rows]
```

```python
olympics_df = olympics_df.dropna(how="any")
olympics_df.isnull().sum()
```

```python
# Load GDP dataframe
gdp_df = pd.read_csv(data_gdp)
gdp_df.head()
```

```python
# Merge GDP column only with olympics_df
olympics_df = olympics_df.merge(gdp_df['GDP'], left_index=True, right_index=True)
olympics_df.head()
```

```python
# Drop extraneous columns that will not be a potential ML feature
olympics_df = olympics_df.drop(['ID', 'Team', 'City', 'Host City', 'Row'], axis=1)

# Rename and reorder the dataframe for preprocessing
olympics_df = olympics_df.rename(columns={'Name': 'Athlete',
                                         'Sex': 'Gender',
                                         'region': 'Region',
                                         'Medal': 'Medal?'})
olympics_df = olympics_df[['Year',
                           'Season', 
                           'Games', 
                           'Host Country', 
                           'NOC', 
                           'Region', 
                           'GDP',
                           'Athlete', 
                           'Gender', 
                           'Age',
                           'Height',
                           'Weight',
                           'Sport', 
                           'Event', 
                           'Medal?', 
                           'Medal_Type']]
olympics_df.head()
```

```python
# Check for null values
olympics_df.isnull().sum()
```

```python
# Export and save dataframe
olympics_df.to_csv('data/olympics.csv', index=False)
```
