```python
# Import our dependencies
import pandas as pd
import numpy as np
```

```python
dfFull = pd.read_csv("data/cleaned_athlete_gdp_sanction_data.csv")
print(dfFull.shape)
dfFull.head()
```

```python
dfFull.nunique()
```

```python
missing = dfFull.loc[(dfFull["Employment"]==0) & (dfFull["UnemploymentRate"]== 0) & (dfFull["GDP"]==0) & (dfFull["Population"])==0]
missing
```

```python
NOC_missing = missing['NOC'].unique()
print(len(NOC_missing))
NOC_missing.tolist()
```

```python
complete = dfFull[(dfFull["Employment"]!=0) & (dfFull["UnemploymentRate"]!= 0) & (dfFull["GDP"]!=0) & (dfFull["Population"])!=0]
complete
```

```python
complete['Year'].unique()
```

```python
NOC_complete = complete['NOC'].unique()
print(len(NOC_complete))
NOC_complete.tolist()
```
