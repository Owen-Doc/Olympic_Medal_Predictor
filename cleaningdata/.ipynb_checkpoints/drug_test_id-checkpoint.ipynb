```python
# Add the Pandas dependency.
import pandas as pd
import numpy as np
import os
import re
from datetime import datetime, date
```

```python
sanctions_file = os.path.join("..","Resources","May-2021-Sanctions-List-Full.xlsx")
sanctions_df = pd.read_excel (sanctions_file, sheet_name='Table 1')
#sanctions_df
athlete_file = os.path.join("..","Resources","cleaned_athlete_data.csv.zip")
athlete_data_df = pd.read_csv(athlete_file)
athlete_data_df
```

```python
DataAge = np.datetime64("2021-07-11")
LifeTimeBan = np.datetime64("2099-12-31")
DataAge
```

```python
#athlete_data_df = athlete_data_df.rename(columns={ 'ID': 'AthleteID', 'Age': 'ParticipatedAge'})
athlete_data_df = athlete_data_df.rename(columns={ 'ID': 'AthleteID'})
sanctions_df = sanctions_df.rename(columns={
    'Name': 'Name',
    'Date of Birth': 'DateofBirth',
    'Nationality': 'NOC',
    'Role': 'Role',
    'Sex': 'Sex',
    'Discipline 1': 'Discipline1',
    'Discipline 2': 'Discipline2',
    'Date of Infraction': 'DateofInfraction',
    'Sanction': 'Sanction',
    'Ineligibility until': 'IneligibilityUntil',
    'Period of Disqualification\nof results': 'PeriodofDisqualificationOfResults',
    'Unnamed: 11': 'DQ2',
    'Lifetime Ban?': 'LifetimeBan',
    'Infraction Type': 'InfractionType',
    'ADRV Rules': 'ADRVRules',
    'ADRV Notes': 'ADRVNotes',
    'Description': 'Description',
    'Description 2': 'Description2'
    })
sanctions_df
```

```python
sanctions_df.dtypes
```

```python
#S = sanctions_df['Name']
#S = S.str.split(pat=", ")
S = sanctions_df["Name"].str.split(", ", n = 1, expand = True)
sanctions_df["AthleteName"]= S[1] + ' ' + S[0]
sanctions_df["FirstName"]= S[1]
sanctions_df["LastName"] = S[0].apply(lambda x: x.split()[-1]).replace({r'[\(\)]': ''}, regex=True)
```

```python
sanctions_df["LastName"]
```

```python
#concatenant and dump second field
sanctions_df['PeriodofDisqualificationOfResults'].update(sanctions_df.pop('DQ2'))
sanctions_df['Description'].update(sanctions_df.pop('Description2'))
#clean leading, trailing and replace newlines with spaces
sanctions_df = sanctions_df.replace({r'\n': ' ', r'\s\s': ' '}, regex=True).replace({r'\s+$': '', r'^\s+': ''}, regex=True)
```

```python
sanctions_df
```

```python
sanctions_df.dtypes
```

```python
sanctions_df["Age"] = (((DataAge - sanctions_df['DateofBirth']) /np.timedelta64(1,'Y')).apply(np.floor)).astype(int)
#sanctions_df = pd.concat([sanctions_df,Age_df],axis=1)
sanctions_df["Age"]
```

```python
# Re-order Columns to move athlete information to the front
cols = list(sanctions_df.columns)
cols = [cols[-1]] + cols[:-1]
cols = [cols[-1]] + cols[:-1]
cols = [cols[-1]] + cols[:-1]
cols = [cols[-1]] + cols[:-1]
cols
```

```python
sanctions_df = sanctions_df[cols]
#sanctions_df
#check with one example
sanctions_df[ (sanctions_df['LastName'] == 'LYSENKO')]
```

```python
sanctions_df.isnull().sum()
```

```python
sanctions_df = sanctions_df.replace(np.nan, '', regex=True)
sanctions_df.isnull().sum()


```

```python
#athlete_data_df['ID']['Name']
FirstName = sanctions_df.FirstName.values.tolist()
LastName = sanctions_df['LastName'].values.tolist()
```

```python
id_df = athlete_data_df.groupby(['AthleteID'])['AthleteID','Name','Sex','NOC'].max() #max to get ages that match the sanction lists if needed
#id_df['ParticipatedAge'] = id_df.ParticipatedAge.astype(int)
#id_df['Year'] = id_df.Year.astype(int)
#id_df['Age'] = 2018 - id_df['Year'] + id_df['ParticipatedAge']
#id_df.drop('ID', axis='columns', inplace=True)
id_df
```

```python
dfFull=[]
dfFull = id_df.merge(sanctions_df, on=['Sex','NOC'])
#sanctions_df.drop('join', axis=1, inplace=True)
#dfFull[ (dfFull['AthleteID'] == 72681)]
dfFull
```

```python
#dfFull['match'] = dfFull.apply(lambda x: (x.Name_x.lower().find(x.FirstName.lower() + ' ') & x.Name_x.lower().find(x.LastName.lower())), axis=1).ge(0)
dfFull['match1'] = dfFull.apply(lambda x: (x.Name_x.lower().find(x.LastName.lower())), axis=1).ge(0) 
dfFull['match2'] = dfFull.apply(lambda x: (x.Name_x.lower().find(x.FirstName.lower() + ' ')), axis=1).ge(0)
```

```python
# match back the first and lastname(maiden name) to update the athlete with sanction data
dfResult = dfFull[(dfFull['match1'] == True) & (dfFull['match2'] == True)].drop(columns={'match1','match2'})
dfResult
```

```python
compression_opts = dict(method='zip', archive_name='SanctionedAthletes.csv')  
dfResult.to_csv('../Resources/SanctionedAthletes.csv.zip', compression=compression_opts)
```
